## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
orders_df= pd.read_csv('dataset/orders_dataset.csv')

orders_df
```

```python
customers_df = pd.read_csv('dataset/customers_dataset.csv')

customers_df
```

Menggabungkan dataset customer dan order berdasarkan customer_id untuk mendapatkan data customer untuk tiap order.

```python
new_order_df = pd.merge(
    left=customers_df,
    right=orders_df,
    how="inner",
    left_on="customer_id",
    right_on="customer_id"
)

new_order_df
```

Menggabungkan dataset detail order, kategori produk, dan translasi kategori produk dalam bahasa Inggris untuk mengetahui kategori produk untuk tiap order.

```python
product_df = pd.read_csv('dataset/order_items_dataset.csv')
category = pd.read_csv('dataset/products_dataset.csv')
productnames_eng = pd.read_csv('dataset/product_category_name_translation.csv')

category_df = pd.merge(category, productnames_eng, on='product_category_name', how = 'outer')

new_product_df = pd.merge(
    left=product_df,
    right=category_df,
    how="inner",
    left_on="product_id",
    right_on="product_id"
)

new_product_df
```

Menggabungkan dataframe orders dan products untuk mendapatkan informasi detail mengenai siapa yang melakukan order, customer melakukan order produk apa dan kapan, dan lainnya.

```python
order_product_df = pd.merge(
    left=new_order_df,
    right=new_product_df,
    how="inner",
    left_on="order_id",
    right_on="order_id"
)

order_product_df
```

Menggabungkan dataframe orders products dengan dataset review tiap order yang dilakukan customer untuk mengetahui tingkat kepuasan customer terhadap order yang telah mereka lakukan.

```python
reviews = pd.read_csv('dataset/order_reviews_dataset.csv')

review_df = pd.merge(
    left=order_product_df,
    right=reviews,
    how="inner",
    left_on="order_id",
    right_on="order_id"
)

review_df
```

Karena jumlah value yang kosong belum mencapai 50% dari total data, row yang mengandung missing value dapat dihapus.

```python
review_df.dropna(axis=0, inplace=True)
review_df
```

```python
review_df.isnull().sum()
```

Selanjutnya, saya akan menghapus beberapa kolom yang tidak terlalu digunakan dalam percobaan ini untuk mengurangi fiturnya.

```python
kolom_dihapus = ['customer_zip_code_prefix', 'product_name_lenght', 'product_description_lenght', 'product_photos_qty', 'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm', 'review_comment_title', 'review_comment_message']

review_df = review_df.drop(columns=kolom_dihapus)

review_df
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Dari mana asal tempat dengan jumlah pelanggan tertinggi yang membeli produk dari e-commerce ini?

```python
state_order_counts = review_df.groupby(by="customer_state").order_id.nunique().sort_values(ascending=False)

# Visualisasi dengan matplotlib
plt.figure(figsize=(10, 6))
state_order_counts.plot(kind='barh', color='skyblue')
plt.title('Jumlah Order per State Customer')
plt.xlabel('Jumlah Order')
plt.ylabel('State Customer')
plt.show()
```

### Pertanyaan 2: Apa saja kategori produk yang paling sering dibeli oleh customer?

```python
plt.figure(figsize=(12, 8))
sns.countplot(x=review_df['product_category_name_english'], order=review_df['product_category_name_english'].value_counts().nlargest(10).index)
plt.title('10 Kategori Produk Teratas yang Dibeli Pelanggan')
plt.xlabel('Kategori Produk')
plt.ylabel('Frekuensi')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

### Pertanyaan 3: Bagaimana tingkat performa pengantaran produk kepada customer?

```python
delivery_counts = review_df['delivery_on_time'].value_counts()
labels = delivery_counts.index.map({True: 'On Time', False: 'Not On Time'})

plt.pie(delivery_counts.values, labels=labels, autopct='%1.1f%%')

# Tampilkan pie chart
plt.title('Delivery Status')
plt.show()
```

### Pertanyaan 4: Kategori produk apa yang memiliki rata-rata nilai review tertinggi dan terendah?

```python
mean_review_df = review_df.groupby("product_category_name_english").review_score.mean().sort_values(ascending=False).reset_index()
mean_review_df.head(15)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))
 
# colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
 
sns.barplot(x="review_score", y="product_category_name_english", data=mean_review_df.head(5), ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Best Performing Product", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)
 
sns.barplot(x="review_score", y="product_category_name_english", data=mean_review_df.sort_values(by='review_score', ascending=True).head(5), ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Worst Performing Product", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)
 
plt.suptitle("Best and Worst Performing Product by Review Score", fontsize=20)
plt.show()
```
